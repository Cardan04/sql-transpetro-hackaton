#  API  Marine Weather API (Open-Meteo)

https://marine-api.open-meteo.com/v1/marine

Esta API foi utilizada para coletar dados ocenicos em tempo real com base em coordenadas martimas precisas. Atravs dela, foi possvel consultar informaes diretamente relacionadas s condies do mar, informando latitude e longitude das regies navegadas, permitindo acompanhar parmetros como altura das ondas, direo do vento, temperatura da superfcie e demais indicadores essenciais para navegao e segurana martima.

### Breve amostra dos dados da API:

```python
import requests

# -----------------------------
# Coordenadas da cidade
# Exemplo: Rio de Janeiro
# -----------------------------
latitude = -22.9068
longitude = -43.1729
date = "2025-01-10"

# -----------------------------
# URL da API pblica
# -----------------------------

url = (
    "https://marine-api.open-meteo.com/v1/marine?"
    f"latitude={latitude}&longitude={longitude}"
    f"&hourly=wave_height,wave_direction,wave_period,windspeed_10m,winddirection_10m"
    f"&start_date={date}&end_date={date}"
)

response = requests.get(url)
data = response.json()

# Se a API no retornar dados
if "hourly" not in data:
    print(" Erro: API no retornou dados")
    print(data)
else:
    print(f" Dados do dia {date}  Rio de Janeiro\n")
    hourly = data["hourly"]

    for i in range(len(hourly["time"])):
        print(f"Hora: {hourly['time'][i]}")
        print(f" Vento: {hourly['windspeed_10m'][i]} m/s  direo {hourly['winddirection_10m'][i]}")
        print(f" Ondas: altura {hourly['wave_height'][i]} m  perodo {hourly['wave_period'][i]} s  direo {hourly['wave_direction'][i]}")
        print("-" * 40)

```

#  API  Open-Meteo Weather Forecast



https://api.open-meteo.com/v1/forecast

Esta API foi empregada para obter previses meteorolgicas das cidades costeiras prximas ao trajeto do navio. A partir dos dados geogrficos das localidades, foi possvel acessar informaes climticas como temperatura, precipitao, ventos e condies atmosfricas gerais recurso til para monitoramento ambiental, logstica martima e planejamento de rota.

### Breve amostra dos dados da API:

```python
import requests

# -----------------------------
# Coordenadas da cidade
# Exemplo: Santos - SP
# -----------------------------
latitude = -23.9608
longitude = -46.3336

# -----------------------------
# URL da API pblica
# -----------------------------
url = f"https://api.open-meteo.com/v1/forecast?" \
      f"latitude={latitude}&longitude={longitude}&" \
      f"hourly=temperature_2m,relativehumidity_2m,wind_speed_10m,winddirection_10m,pressure_msl,precipitation"

# -----------------------------
# Requisio HTTP
# -----------------------------
resposta = requests.get(url).json()

# -----------------------------
# Exibir dados da primeira hora
# -----------------------------
print(" Local consultado:", latitude, longitude)
print(" Temperatura:", resposta["hourly"]["temperature_2m"][0], "C")
print(" Umidade:", resposta["hourly"]["relativehumidity_2m"][0], "%")
print(" Vento:", resposta["hourly"]["wind_speed_10m"][0], "m/s")
print(" Direo do vento:", resposta["hourly"]["winddirection_10m"][0], "")
print(" Presso:", resposta["hourly"]["pressure_msl"][0], "hPa")
print(" Precipitao:", resposta["hourly"]["precipitation"][0], "mm")
```
